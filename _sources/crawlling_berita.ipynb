```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
datas=[]

url = 'https://www.jawapos.com/berita-hari-ini/'
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
berita = soup.findAll('div', 'post-list__item')
for item in berita:
  link = item.find('a','d-block')['href']
  judul = item.find('h3','post-list__title').text
  req2 = requests.get(link)
  soup2 = BeautifulSoup(req2.text, 'html.parser')
  try:
    isi = soup2.find('div', 'content').text.replace('\n', '')
  except:
    isi=''
  datas.append([judul, isi])




```

```python
df = pd.DataFrame(datas, columns=['judul', 'isi'])
df
```

```python
df.to_csv('crawl_berita.csv')
```
