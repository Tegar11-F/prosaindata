## Graph 

```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4),(4,5),(4,6),(5,6),(5,7),(5,8),(6,7),(6,8),(7,8),(7,9)])
```

```python
nx.draw(G, with_labels=True)

```

```python
nx.degree_centrality(G) #banyaknya sebaran
```

```python
nx.closeness_centrality(G) #mencari node terdekat
```

```python
nx.betweenness_centrality(G, k=None, normalized=True, weight=None, endpoints=False, seed=None)
```

```python
pr = nx.pagerank(G) #alpha default 0.8
```
